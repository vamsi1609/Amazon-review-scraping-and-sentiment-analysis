## What is Scrapy?

[Scrapy](https://scrapy.org/) is a ***web crawling framework*** for a developer to write code to create, which defines how a particular site (or a group of websites) will be scrapped. The most significant feature is that it is built on Twisted, an asynchronous networking library, which makes the spider performance is very significant.

```
%%writefile /content/amazon_reviews_scraping/amazon_reviews_scraping/amazon_reviews.py
import scrapy
# Loading the phones list and corresponding urls of reviews
import pandas as pd
df=pd.read_csv('/content/phone_list.csv')
product=df['Product-name'].copy()
URL = df['URL'].copy()
# Creating a new class to implement Spide
class AmazonReviewsSpider(scrapy.Spider):
    start_urls=[]
    # Spider name
    name = 'amazon_reviews'
 
    # Domain names to scrape
    allowed_domains = ['amazon.in']
    # Base URL for the MacBook air reviews
 
    # Creating list of urls to be scraped by appending page number a the end of base url
    for i in range(10):
       start_urls.append(URL[i]+'1')
       print(start_urls[i])
 
    # Defining a Scrapy parser
    def parse(self, response):
              data = response.css('#cm_cr-review_list')
 
            # Collecting product star ratings
              star_rating = data.css('.review-rating')
 
            # Collecting user reviews
              comments = data.css('.review-text')

              data = response.css('#a-page')

            # Collecting phone name 
              phone_name = data.css('.a-size-large.a-text-ellipsis')
              count = 0
 
            # Combining the results
              for review in star_rating:
                  yield{
                      'Model-Name': ''.join(phone_name.xpath('.//text()').extract()),
                      'comment': ''.join(comments[count].xpath(".//text()").extract()),
                      'stars': ''.join(review.xpath('.//text()').extract())
                       }
                  count=count+1
```
